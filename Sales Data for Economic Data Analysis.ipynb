```python
# Convert 'Year' column from float to integer
df['Year'] = df['Year'].astype(int)
```

```python
# Round the 'Unit Cost' column to whole numbers
df['Unit Cost'] = df['Unit Cost'].round(0)
```

```python
# Convert 'Unit Cost' column from float to integer
df['Unit Cost'] = df['Unit Cost'].astype(int)
```

```python
# Round the 'Unit Price' column to whole numbers
df['Unit Price'] = df['Unit Price'].round(0)
```

```python
# Calculate Profit and Profit Margin
df['Profit'] = df['Revenue'] - df['Cost']
df['Profit Margin'] = df['Profit'] / df['Revenue'] * 100

# Check the new columns
df.head()
```

```python
df.info()
```

```python
df['Profit Margin'] = df['Profit Margin'].round(0)
```

```python
df['Profit Margin'] = df['Profit Margin'].astype(int)
```

```python
df.head()
```

```python
# To view the profit margin for each product category
profit_margin_by_category = df.groupby('Product Category')['Profit Margin'].mean().reset_index()
(profit_margin_by_category)
```

```python
#Plot a bar chart for Average Profit Margin by Product Category
plt.figure(figsize=(8, 6))
sns.barplot(data=profit_margin_by_category, x='Product Category', y='Profit Margin', palette='Blues')
plt.title('Average Profit Margin by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Profit Margin (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sum Profit by Product Category
profit_by_category = df.groupby('Product Category')['Profit'].sum()

# Plot the profit by product category
profit_by_category.plot(kind='bar', title='Profit by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Profit')
plt.show()
```

```python
# Group by 'Product Category' and sum the 'Profit'
product_profit = df.groupby('Product Category')['Profit'].sum().reset_index()

# Create the bar plot
g = sns.barplot(data=product_profit, x='Product Category', y='Profit')

# Annotate the bars with the profit values
for v in product_profit.itertuples():
    g.text(x=v.Index, y=v.Profit + 1e4, s=str(v.Profit), size=10, ha="center")

# Rotate x-axis labels for better readability
plt.xticks(rotation=60)
plt.title('Profit by Product Category')
plt.show()
```

```python
# Calculate the gender distribution
gender_counts = df['Customer Gender'].value_counts()

# Define a bluish color palette
colors = ['#1f77b4', '#17becf']  

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Customer Gender Distribution')
plt.axis('equal')  
plt.show()
```

```python
# Group by 'Country' and sum the 'Profit'
country_profit = df.groupby('Country')['Profit'].sum().reset_index()

# Create the bar plot
g = sns.barplot(data=country_profit, x='Country', y='Profit')
# Rotate x-axis labels for better readability
plt.xticks(rotation=60)
plt.title('Profit by Country')
plt.show()
```

```python
# Top 5 best-selling products
df.groupby('Product Category')['Quantity'].sum().sort_values(ascending=False).head(5)
```

```python
# Group by subcategory and sum the profits
subcategory_profit = df.groupby('Sub Category')['Profit'].sum().reset_index()

# Sort the data by profit
subcategory_profit = subcategory_profit.sort_values(by='Profit', ascending=False)

# Visualize the results with a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=subcategory_profit, x='Sub Category', y='Profit', palette='Blues')
plt.title('Profit by Subcategory')
plt.xticks(rotation=45, ha='right')  # Rotate x labels for better readability
plt.show()
```

```python
# Group by country and product category, summing the profits
country_category_profit = df.groupby(['Country', 'Product Category'])['Profit'].sum().reset_index()

# Find the product category with the maximum profit for each country
max_profit_per_country = country_category_profit.loc[country_category_profit.groupby('Country')['Profit'].idxmax()]

plt.figure(figsize=(8, 6))
sns.barplot(data=max_profit_per_country, x='Country', y='Profit', hue='Product Category', palette='Blues')
plt.title('Product Category with Maximum Profit by Country')
plt.xticks(rotation=45, ha='right')  
plt.legend(title='Product Category')
plt.show()
```

```python
# Create a barplot for over all profit per year
overall_profit_per_year = df.groupby('Year')['Profit'].sum().reset_index()

plt.figure(figsize=(7, 6))
sns.barplot(data=overall_profit_per_year, x='Year', y='Profit', color='skyblue')
plt.title('Overall Profit per Year')
plt.xlabel('Year')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.show()
```

```python
# Top 5 best-selling products
df.groupby('Product Category')['Quantity'].sum().sort_values(ascending=False).head(5)

# Profit Margin by State
profit_by_country = df.groupby('Country')['Profit Margin'].mean().sort_values(ascending=False)
profit_by_country.plot(kind='bar', title='Average Profit Margin by Country')
plt.show()
```

```python
# Create a count plot for ptofit performance by country per year
profit_per_country_year = df.groupby(['Country', 'Year'])['Profit'].sum().reset_index()

plt.figure(figsize=(10, 8))
sns.barplot(data=profit_per_country_year, x='Year', y='Profit', hue='Country', palette='Blues')
plt.title('Profit Performance by Country per Year')
plt.xlabel('Year')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend outside
plt.show()
```

```python
# Create a count plot for product category by customer gender
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Product Category', hue='Customer Gender', palette='Blues')
plt.title('Product Purchases by Customer Gender')
plt.xlabel('Product Category')
plt.ylabel('Count of Purchases')
plt.xticks(rotation=45)  
plt.legend(title='Customer Gender')  
plt.show()
```

```python
# Define age groups
bins = [0, 18, 25, 35, 45, 55, 65, 100]  # Define age bins
labels = ['0-18', '19-25', '26-35', '36-45', '46-55', '56-65', '66+']  # Labels for age bins
df['Age Group'] = pd.cut(df['Customer Age'], bins=bins, labels=labels, right=False)
```

```python
age_country_profit = df.groupby(['Country', 'Age Group'])['Profit'].sum().reset_index()
```

```python
# Create a bar plot for Total Profit by Age Group and Country
plt.figure(figsize=(10, 6))
sns.barplot(data=age_country_profit, x='Age Group', y='Profit', hue='Country', palette='Blues_d')  # Use 'Blues_d' for darker colors
plt.title('Total Profit by Age Group and Country')
plt.xlabel('Age Group')  
plt.ylabel('Total Profit')
plt.xticks(rotation=45)  
plt.legend(title='Country')  
plt.show()
```

```python
gender_country_profit = df.groupby(['Country', 'Customer Gender'])['Profit'].sum().reset_index()
```

```python
#  Create a bar plot for total profit by gender for each country
plt.figure(figsize=(10, 6))
sns.barplot(data=gender_country_profit, x='Customer Gender', y='Profit', hue='Country', palette='Blues_d')  # Use 'Blues_d' for darker colors
plt.title('Total Profit by Customer Gender and Country')
plt.xlabel('Customer Gender')  # Gender on x-axis
plt.ylabel('Total Profit')
plt.xticks(rotation=45)  
plt.legend(title='Country')  
plt.show()
```
