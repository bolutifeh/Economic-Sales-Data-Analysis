```python
#importing the SQLite3 module for working with SQLite databases and the Pandas library for data manipulation and analysis.

import sqlite3
import pandas as pd
```

```python
df = pd.read_csv('cleaned_sales_data.csv')
```

```python
# Connect to a database (will create if not exists)
conn = sqlite3.connect('sales_data.db')
cursor = conn.cursor()

# Write records stored in a DataFrame to a SQL database
df.to_sql('sales', conn, if_exists='replace', index=False)
```

```python
total_sale_per_year = """
SELECT "Year", SUM(Profit) AS Total_Profit
FROM sales
GROUP BY "Year"
ORDER BY Total_Profit DESC;
"""

# Execute the SQL query and store the result in a Pandas DataFrame
total_sale_per_year = pd.read_sql_query(total_sale_per_year, conn)

# Display the DataFrame containing the query result
total_sale_per_year
```

```python
# Define the product category you want to filter by
product_category = 'Clothing' 

filtered_transactions_query = f"""
SELECT *
FROM sales
WHERE "Product Category" = '{product_category}';
"""

# Execute the query
filtered_transactions = pd.read_sql_query(filtered_transactions_query, conn)

# Print the results
(filtered_transactions)
```

```python
# Query to get the top 10 best-selling products by total sales revenue
top_selling_products_query = """
SELECT "Sub Category", SUM(Revenue) AS Total_Sales
FROM sales
GROUP BY "Sub Category"
ORDER BY Total_Sales DESC
LIMIT 10;
"""

# Execute the query and store the result in a DataFrame
top_selling_products = pd.read_sql_query(top_selling_products_query, conn)

# Display the results
(top_selling_products)
```

```python
# Query to get the top 10 best-selling products by total profit
top_selling_products_query = """
SELECT "Sub Category", SUM(Profit) AS Total_Profit
FROM sales
GROUP BY "Sub Category"
ORDER BY Total_Profit DESC
LIMIT 10;
"""

# Execute the query and store the result in a DataFrame
top_selling_products = pd.read_sql_query(top_selling_products_query, conn)

# Display the results
(top_selling_products)
```

```python
# Query to calculate total revenue by customer gender
total_revenue_by_gender_query = """
SELECT "Customer Gender", SUM(Revenue) AS Total_Revenue
FROM sales
GROUP BY "Customer Gender"
ORDER BY Total_Revenue DESC;
"""

# Execute the query and store the result in a DataFrame
total_revenue_by_gender = pd.read_sql_query(total_revenue_by_gender_query, conn)

# Display the results
(total_revenue_by_gender)
```
